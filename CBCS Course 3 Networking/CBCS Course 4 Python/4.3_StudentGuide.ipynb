## Introduction

So far we've observed variables that hold single pieces of data. Lists, on the other hand, can hold multiple pieces of data within a variable. 

A list in Python is a collection of elements put into an array. A list element can be of any data type, such as strings, integers, floats and even other lists. The following are some examples of lists:

```
Numbers = [10, 15, 3.4, 5]
Items = [“coat”, “shoes”, “socks”]
Info = [“they”, 22, “new york”]
All_info = [“they”, 22, “new york”], [“she”, 28, “quito”]
```

Each element in the array gets an index. This can be referred to later, just as we did in strings.
The way we create a list is by using the squared brackets and using commas to separate each object.

````python
List_1 = [1,2,3,4]

print(List_1)
````

Just like strings, lists also have methods that are unique to them and help us manipulate, edit and print them.

Here are some examples of list methods:

`append()`: Adds an element at the end of the list

`clear()`: Removes all elements from the list 

`count()`: Returns the number of elements with the specified value

`index()`: Returns the index of the specified value

`remove()`: Removes the first item with the specified value 

`pop()`: Removes the element with the specified position form the list 

`insert()`: Inserts an element at the specified position

`reverse()`: Reverses the list

`len()`: Get the length of the list 

`sort()`: Sort the list alphabetically

# Activity 3B: List Indexing

## Accessing List Items by Index
Just like string, lists also uses indexes. Each index maps to one of the values. Also like strings, the first element of a list is always at index zero. You can access individual items in a list using an index in square brackets.

````python
L = ['red', 'green', 'blue', 'yellow', 'black']
````

````python
print(L[0])
````

````python
print(L[2])
````

Python will raise an `IndexError` if you use an index that exceeds the number of items in your list.

````python
L = ['red', 'green', 'blue', 'yellow', 'black']
print(L[10])
````

## Negative List Indexing
You can access a list by negative indexing as well. Negative indexes count backward from the end of the list. So, `L[-1]` refers to the last item, `L[-2]` is the second-last, and so on.

````python
L = ['red', 'green', 'blue', 'yellow', 'black']
print(L[-1])
````

````python
print(L[-2])
````

# Activity 3C: Indexing Logs
## Introduction
Earlier we used a method called `split()` that takes a string and creates a list.

````python
list_1 = "this is a string".split()
print(list_1)
````

Indexing is very useful for accessing different elements or values in a string, like in an Apache log line.

Let's take a look at this log line:

`Log_line = '10.0.2.17 - - [19/Feb/2022:13:57:58 -0500] "GET / HTTP/1.1" 200 10956 "-" "curl/7.68.0"'`

If you relate to this as a string, it will be difficult to access different items because you will need to know the exact start and end indexing. If you use `split()`, you can create a list of each element in this log line.

````python
Log_line = '10.0.2.17 - - [19/Feb/2022:13:57:58 -0500] "GET / HTTP/1.1" 200 10956 "-" "curl/7.68.0"'
list_1 = Log_line.split()
print(list_1)
````

As you can see, there are a couple of elements of interest to us. For example,
the `source_ip` is at the first element.

````python
print(list_1[0])
````

Another element is the timestamp, which is the fourth element.

````python
print(list_1[3])
````

# Activity 3D: List Slicing
## Introduction

Just like strings, we can slice a list according to indexes.

````python
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
````

Following are 3 numbers in the brackets; the start of the slice (inclusive), the end of the slice (exclusive) and the step.

````python
print(L[2:7:2])
````

This prints every second item from the third element (inclusive) to the seventh element (exclusive).

## Slice with Negative Indices

````python
print(L[-7:-2])
````

## Slice at the Beginning and End

Omitting the start index will start the slice from the index 0. This means that `L[:stop]` is equivalent to `L[0:stop]`

In the following example, we slice the first three items from the list. 

````python
L = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
print(L[:3])
````

Omitting the stop index extends the slice to the end of the list. This means that `L[start:]` is equivalent to 
`L[start:len(L)]`

In the following example, we slice everything after the sixth element.

````python
print(L[6:])
````

## Slicing Every Second Item 

The step indicates how many elements are skipped. In the following example, we slice every second item in the list.

````python
print(L[::2])
````

Similarly, the following will print every third element.

````python
print(L[::3])
````

## Reverse a List

You can reverse a list by omitting both start and stop indices and specifying a step as -1.

````python
L = ['a', 'b', 'c', 'd', 'e']
print(L[::-1])
````

# Activity 3E: Lists - Putting It All Together
## Introduction

Let's use these methods to read a line from a log, `auth.log`. This log is used to log any authentication attempts on a Linux machine.

When we have a successful login, the line should look like the following.

````python
Log_line1 = "Apr 15 11:06:04 kali sshd[8350]: Accepted password for kali from 10.0.0.1 port 60194 ssh2"
````

The information we are interested in is the date, time, username, source IP and machine name.

At this point, the variable is a string data type. Using the string `split()`
method, we can split this string into items in a list. Note that the delimiter, unless mentioned otherwise, is a space.

````python
Array_1 = Log_line1.split()

print(Array_1)
````

We can use this logic to parse this line to get an output that prints the information we're interested in.

The last three elements are not useful to us at this time, so let's slice them out.

````python
date = Array_1[0:2]
print(date)
````

We got a list! For convenience, let's create a string. We can use the `join()` method, which takes a list of items and creates a string. This is the opposite of `split()`. First, we need to decide what will separate the string (usually spaces). Then in the parentheses, we put the list.

````python
Date = "".join(date)
print(Date)
````

To get the time we need the 3rd element.

````python
Time = Array_1[2]
print(Time)
````

Similarly we can find the rest of the items in this list.

````python
Time = Array_1[2]

Machine_name = Array_1[3]

User_name = Array_1[3]

Source_ip = Array_1[6]
````

Now lets print the result.

````python
print("The User "+User_name+" "+"logged on "+Date+" "+" at "+Time +" from "+Source_ip+" "+"to "+Machine_name)
````

Let’s put it all together.

````python
Log_line1 = "Apr 15 11:06:04 kali sshd[8350]: Accepted password for kali from 10.0.0.1 port 60194 ssh2"

Array_1 = Log_line1.split()
date = Array_1[0:2]
Date = "".join(date)
Time = Array_1[2]
Machine_name = Array_1[8]
User_name = Array_1[3]
Source_ip = Array_1[10] 
print("The User "+User_name+" "+"logged on "+Date+" "+"at "+Time+" from "+Source_ip+" "+"to "+Machine_name)
````
